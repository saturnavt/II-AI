```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Load csv
data = pd.read_csv('resources/DATOS_DE_CALIDAD_DEL_AIRE_EN_COLOMBIA_2011-2018.csv', low_memory = False)
data['Fecha'] = data['Fecha'].str[:-15]
```

```python
envigado = data[data.Nombre_del_municipio == 'ENVIGADO']
data.sample(30)
```

```python
plt.plot(envigado.Fecha.sample(3), envigado.Concentracion.sample(3))
plt.title("Calidad de aire en envigado")
plt.xlabel("Año")
plt.ylabel("Concentración")
plt.legend(["Concentración"])
plt.show()
```
