```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('resources/growOfLanguageByYear.csv')
```

```python
data
```

```python
python = data[data.language == 'python']
r = data[data.language == 'R']
cpp = data[data.language == 'c++']
java = data[data.language == 'java']
javascript = data[data.language == 'javascript']

plt.plot(python.year, python.porcentage)
plt.plot(r.year, r.porcentage)
plt.plot(cpp.year, cpp.porcentage)
plt.plot(java.year, java.porcentage)
plt.plot(java.year, java.porcentage)
plt.plot(javascript.year, javascript.porcentage)

plt.title("most used languages for AI by year")
plt.xlabel("Porcentage")
plt.ylabel("Year")
plt.legend(["Python", "R", "C++", "java", "JavaScript"])
plt.show()
```
