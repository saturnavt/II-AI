```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#PYTHON
x = [1995, 2006, 2021]
y = [1, 4, 9]
#R
xR = [1995, 2006, 2021]
yR = [1, 9, 0]
#C++
xc = [1995, 2006, 2021]
yc = [1, 8, 8]
#Java
xj = [1995, 2006, 2021]
yj = [1, 7, 7]
#JS
xjs = [1995, 2006, 2021]
yjs = [1, 4, 3]

#PYTHON PLOT
plt.plot(x, y)
#R PLOT
plt.plot(xR, yR)
#C++ PLOT
plt.plot(xc, yc)
#Java PLOT
plt.plot(xj, yj)
#JS PLOT
plt.plot(xjs, yjs)

plt.title("most used languages for AI by year")
plt.xlabel("Porcentage")
plt.ylabel("Year")
plt.legend(["Python", "R"])
plt.show()
```
